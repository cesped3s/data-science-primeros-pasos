# Question 5: Plot Tesla Stock Graph

```python
def make_graph(data, title):
    plt.figure(figsize=(18, 6))
    plt.plot(data['Close'], label='Precio de Cierre', color='blue')
    plt.title(title)
    plt.xlabel('Fecha')
    plt.ylabel('Precio de Cierre ($)')
    plt.legend()
    plt.grid(True)
    plt.show()  # Esto asegura que el gráfico se muestre

# Crear el gráfico con un título
make_graph(tesla_data, 'Precio de Cierre de Tesla (Últimos 6 meses)')
```
